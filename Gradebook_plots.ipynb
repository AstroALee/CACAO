```python

# Technically the gradebook has "numQuestions" but having this number nearby might help with grade boundaries 
plotting_dict = { "numPoints":12 , "showMean":True, "showMedian":True , "showMax" : True , "showMode" : False , 
            "showLetters":True , "showCumulative_bars" : True , "showCumulative_scores" : False , "Cumulative_score_color" : cur_cMap(0.3),
            "ypadding_fac" : 1.2, "lettergrade_fs": 20, "cumulative_ps" : 20 ,
            "xtick_shift" : 0.5 } 

# Column name
cname = 'Course Total'

# Letter grade boundaries, e.g., "A":8 means a score of 8,9,... is an A. Must be in descending 
letter_boundaries = { "A":10 , "B":8 , "C":6, "F":0 }

# Title 
plot_title = "Physics 110, Spring 2025, Quiz 1"
plot_title = f"{class_data.metadata['class_name']}, {class_data.metadata['semester']}, {cname}"

# Locations of left side of bins for histogram + (numPoints+1)
# If every point posssibility gets a bin: np.arange(0,plotting_dict["numPoints"]+1.1,1)
bins = list(np.arange(0,13,1)) #+ [101]

# Optional tickmarks
#draw_xticks = np.arange(0,plotting_dict["numPoints"]+1,1) # for all possible
draw_xticks = np.arange(0,13,1)
draw_yticks = []

# Data to histo 
scores = (gbook[cname].to_numpy()).copy()

# output file name (optional)
outfile = None
outfile = class_data.metadata["output_plot_header"] + f"{cname.replace(' ','_')}" + class_data.metadata["output_plot_footer"]

# Create histogram 
aplt.gen_IntegerHisto(scores,plotting_dict["numPoints"],plotting_dict,letter_boundaries,plot_title,bins,[draw_xticks,draw_yticks],cur_cMap,outfile)

#aplt.gen_QuizCorrections(plotting_dict,0.4,letter_boundaries,0.4,plot_title,0.25,'./Phy110_S25_Quiz2_Correction.png')
```
